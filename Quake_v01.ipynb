```python
%matplotlib ipympl
import matplotlib.pyplot as plt
import matplotlib as mpl 
import pandas as pd
import seaborn as sns
import numpy as np

from matplotlib.colors import ListedColormap
from sklearn.linear_model import LinearRegression

plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

mydata = pd.read_csv('Quake.csv')
```

```python
x=mydata["Lat"]
y=mydata["Long"]
z=mydata["Depthneg"]
s=mydata["Cat"]

# Create a 3D surface plot with Seaborn
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatt = ax.scatter3D(x, y, z, c=z, marker=".", s=2, cmap='PuRd')
ax.set_xlabel('Lat')
ax.set_ylabel('Long')
ax.set_zlabel('Depth')
ax.set_title('Location Depth Plot')

fig.colorbar(scatt,ax=ax, orientation='vertical', label='Depth')



plt.show()
```
